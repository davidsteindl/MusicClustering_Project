```python
import pandas as pd
from IPython.display import display, Markdown
from src.evaluation_utils import plot_songsPerCluster

df = pd.read_csv("../data/processed_data/spotify_clustered.csv")

display(Markdown("### Overview of Songs per Clusters for Main Dataset with 6 Clusters:"))
plot_songsPerCluster(df, "cluster_main")
```

```python
from IPython.display import display, Markdown

display(Markdown("### Mean feature values per cluster for main dataset:"))

display(Markdown("**Main Dataset with 6 clusters:**"))
main_features = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                 'instrumentalness', 'valence', 'tempo', 'duration_ms']

cluster_summary_main = df.groupby("cluster_main")[main_features].mean().round(2)
display(cluster_summary_main)
```
